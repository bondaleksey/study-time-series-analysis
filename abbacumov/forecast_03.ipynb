Когда временной ряд лучше описывать аддитивной моделью, а когда мультипликативной?

Ответ:
Если ряд колеблется внутри полоски вокруг тренда, то ряд лучше описывается аддитивной моделью. Полоска имеет одинаковую ширину.

Если на графике ряд колеблется касаясь сторон угла или нескольких последовательных углов, то ряд лучше описывается мультипликативной моделью.
