Сезонная составляющая присутствует, когда какие-то элементы ряда повторяются через *одинаковые* промежутки времени. Этот промежуток времени называется - **период сезонных отклонений**.

Сезонность - сама по себе временной ряд. Значениями ряда сезонности будут сезонные отклонения ряда от тренда.

Вопрос: "Значения сезонных отклонений меняются ли год от года?"

При прогнозировании ответ зависит от модели, которую использует аналитик.
При `экспоненциальном сглаживании` или `ARIMA` - сезонные поправки меняются, так как эти методы гибкие и адаптивные, но модели сложнее.

При `сезонной декомпозиции` или `регрессионной модели` сезонные поправки не меняются (годовые изменения зашиты в модель).

Вопрос: "Может ли сезонная составляющая принимать отрицательные значения?" 

В аддитивной модели:
$$ X_t = T_t + S_t + E_t $$
отрицательные значения встречаются почти всегда. (Сезонная составляющая измеряет отклонение от тренда)
В годе в точке пика значения (например, продаж) отклонение положительное.
В годе в точке минимума значения (например, продаж) отклонение ниже тренда, а значит отрицательное.

В мультипликативной модели:
$$ X_t = T_t \cdot S_t \cdot E_t $$
 теоретически отрицательные значения могут быть, но на практике такое не встречается.

Обычно период очевиден:
- Если данные месячные, то период равен году.
- Если данные часовые, то период может быть суточным.
- Еще он может быть недельным или 
- квартальным.



## Ошибки 
Аддитивная модель:
$$ E_t = X_t - T_t - S_t$$
Мультипликативная модель:
$$ E_t = \frac{X_t}{T_t \cdot S_t }$$
Мы стараемся так определить модель, чтобы ошибки вели себя как белый шум = последовательность независимых наблюдений случайной величины (не обязательно нормально распределенных, нет автокорреляции = 0)

Если в ошибках возникают закономерности, то аналитик "не доработал" с моделью. Модель можно усовершенствовать, если учесть закономерности в ошибках. 

### Как прогнозируем ошибки
- Если модель аддитивная, прогноз ошибки равен нулю.
- Если модель мультипликативная, прогноз ошибки равен единице.
